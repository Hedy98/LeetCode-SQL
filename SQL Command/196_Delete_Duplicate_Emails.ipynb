**SOLUTION 1: MY SQL (JOIN)**

```python
DELETE p1
FROM person p1
JOIN person p2
  ON p1.email=p2.email AND p1.id > p2.id
```

**SOLUTION 2: MY SQL (SUBQUERY)**

```python
DELETE FROM person
WHERE id NOT IN (
    SELECT *
    FROM (
        SELECT MIN(id) AS id
        FROM person
        GROUP BY email
    ) tb1
)
```
