**181. Employees Earning More Than Their Managers** (Join-Easy)

image.png

**SOLUTION: MY SQL SERVER**

```python
SELECT a.Name as Employee
FROM Employee a
JOIN Employee b
    ON a.managerId = b.Id
WHERE a.salary > b.salary
```
