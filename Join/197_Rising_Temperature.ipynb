**SOLUTION: MY SQL SERVER**

```python
SELECT w2.Id
FROM Weather w1
JOIN Weather w2
    ON DATEADD(day,1,w1.recorddate) = w2.recorddate
WHERE w1.Temperature < w2.Temperature
```
