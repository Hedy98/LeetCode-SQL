**SOLUTION: MY SQL SERVER**

```python
SELECT d.Name as Department, e.Name as Employee, e.Salary
FROM Employee e
    JOIN Department d
    ON e.DepartmentId = d.id
    JOIN (
        SELECT MAX (Salary) as Salary, DepartmentId
        FROM Employee
        GROUP BY DepartmentId
    ) AS mx
    ON e.Salary = mx.Salary and e.DepartmentId = mx.Department.id
```
